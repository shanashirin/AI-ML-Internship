```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.DataFrame({
    'Days':[1,2,3,4],
    'product A':[10,20,30,40],
    'product B':[15,25,35,45]
})
sns.lineplot(x='Days',y='product A',data=data,color='red',marker='o')
sns.lineplot(x='Days',y='product B',data=data,color='blue',marker='o')
sns.set_style("whitegrid")
plt.show()
```
