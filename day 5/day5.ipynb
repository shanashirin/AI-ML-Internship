```python
students=[
    {"name": "Alice", "marks": 85},
    {"name": "Bob", "marks": 92}
]
total=0
for s in students:
    total+=s["marks"]
average=total/len(students)
print("Average marks of the students is:",average)
```

```python
highest=max(marks)
print("Highest marks among the students is:",highest)
```

```python
for i in range(len(students)):
    if marks[i]>75:
        print(students[i],"has scored above 75 marks.")
```

```python
s=input("Enter a string:")
freq={}
for ch in s:
    if ch in freq:
        freq[ch]+=1
    else:
        freq[ch]=1
print("Frequency of each character in the string is:",freq)
```
