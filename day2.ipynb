```python
n=int (input("Enter a number:"))
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,n):
        if n%i == 0:
            return False
    return True
if is_prime(n):
    print("Prime number")
else:
    print("Not a prime number")
```

```python
def fact(n):
    if n<0:
        return "Not defined"
    result=1
    for i in range(1,n+1):
        result*=i
    return result
n=int(input("Enter a number:"))
print("factorial=",fact(n))
```
