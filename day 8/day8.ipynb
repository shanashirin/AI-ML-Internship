```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
```

```python
def prime(n):
    if n>1:
        for i in range(2,n):
            if n%i==0:
                return "not prime"
        return "prime"
print(prime(11))
```

```python
def rev(s):
    return s[::-1]
print(rev("python"))
```

```python
def avg(n):
    return sum(n)/len(n)
print(avg([10,20,30,40,50]))
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
print(factorial(0))
```

```python
def sum(n):
    if n==1:
        return 1
    return n+sum(n-1)
print(sum(5))
```

```python
def fibanocci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibanocci(n-1)+fibanocci(n-2)
print(fibanocci(10))
```

```python
def sum(n):
    if n==0:
        return 0
    return n%10+sum(n//10)
print(sum(12345))
```

```python
def flatten_list(n):
    result=[]
    for i in n:
        if type(i)==list:
            result.extend(flatten_list(i))
        else:
            result.append(i)
    return result
print(flatten_list([1,2,[3,4],[5,6],7]))
```

```python
def count_occurrence(s):
    count={}
    for i in s:
        if i in count:
            count[i]+=1
        else:
            count[i]=1
    return count
print(count_occurrence("hello world"))
```
