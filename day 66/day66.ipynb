```python
# 🔹 FAISS

# FAISS (Facebook AI Similarity Search) is a library used for efficient similarity search and clustering of dense vectors.
# It is widely used in RAG systems to quickly retrieve similar documents.

# 🔹 ChromaDB

# ChromaDB is an open-source vector database designed for storing and retrieving embeddings.
# It is easy to use and commonly used in small to medium RAG applications.

# 🔹 Pinecone

# Pinecone is a cloud-based vector database that provides fast and scalable similarity search.
# It is mainly used in production-level RAG systems.
```

```python
documents=[
    "Ai internship fee is rs 5000",
    "Duration is 6 months",
    "Location is Technopark"
]
question=input("Ask:")
for doc in documents:
    if "fee" in question.lower() and "Fee" in doc:
        print(doc)
```
