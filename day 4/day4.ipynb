```python
try:
    a=int(input("Enter a number:"))
    b=int(input("Enter divisor:"))
    c=a/b
except ZeroDivisionError:
    print("cannot divide by zero")
except ValueError:
    print("Invalid input. Please enter a valid number.")
else:
    print("Result:",c)
finally:
    print("Execution completed.")
```

```python
file=open("data.txt","r")
content=file.read()
words=content.split()
longest=words[0]
for w in words:
    if len(w)>len(longest):
        longest=w
print("Longest word in the file:",longest)
file.close()
```

```python
file=open("data.txt","r")
content=file.read()
count=0
for c in content:
    if c.lower() in "aeiou":
        count+=1
print("Number of vowels in the file:",count)
```

```python
file=open("data.txt","r")
content=file.readlines()
file.close()
unique_lines=list(set(content))
file=open("data.txt","w")   
file.writelines(unique_lines)
file.close()
print("Duplicate lines removed successfully.")
```
