```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
data={
    "student": [1,2,3,4,5],
    "results":[0, 1, 0, 1, 0]

}
df=pd.DataFrame(data)
print(df)
X=df[["student"]]
y=df["results"]
model=DecisionTreeClassifier()
model.fit(X,y)
prediction=model.predict([[1]])
print(prediction)
```

```python
data={
    "loan":[1000, 2000, 1500, 3000, 2500],
    "approval":[0, 1, 0, 1, 0]
}
df=pd.DataFrame(data)
print(df)
X=df[["loan"]]
y=df["approval"]
model=DecisionTreeClassifier()
model.fit(X,y)
prediction=model.predict([[2000]])
print(prediction)
```

```python
data={
    "disease":[0, 1, 0, 1, 0],
    "symptom":[0, 1, 0, 1, 0]
}
df=pd.DataFrame(data)
print(df)
```

```python
X=df[["disease"]]
y=df["symptom"]
model=DecisionTreeClassifier()
model.fit(X,y)
prediction=model.predict([[1]])
print(prediction)
```
