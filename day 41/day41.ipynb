```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
data={
    "Student hours":[1,2,3,4,5],
    "Result":[0,1,1,1,0]
}
df=pd.DataFrame(data)
print(df)
X=df[["Student hours"]]
y=df[["Result"]]
model=RandomForestClassifier()
model.fit(X,y)
print(model.predict([[6]]))
```

```python
data={
    "disease":[0,1,0,1,0],
    "symptom":[0,1,1,1,0]
}
df=pd.DataFrame(data)
print(df)
X=df[["symptom"]]
y=df[["disease"]]
model=RandomForestClassifier()
model.fit(X,y)
print(model.predict([[1]]))
```

```python
data={
    "Loan Amount":[1000,2000,3000,4000,5000],
    "approval":[0,1,1,1,0]
}
df=pd.DataFrame(data)
print(df)
```

```python
X=df[["Loan Amount"]]
y=df[["approval"]]
model=RandomForestClassifier()
model.fit(X,y)
print(model.predict([[6000]]))
```
