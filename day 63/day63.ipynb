```python
training_data={
    "hi":"Greeting",
    "hello":"Greeting",
    "good morning": "Greeting",
    "bye": "Goodbye",
    "track my order": "Order Tracking",
    "where is my order": "Order Tracking",
    "cancel my booking": "Booking Cancellation",
    "refund": "Refund",
    "fees": "Fee Enquiry",
    "admission": "Admission Enquiry"

}
user=input("You:").lower()
found=False
for sentence,intent in training_data.items():
    if sentence in user:
        print("predicted intent:",intent)
        found=True
if not found:
    print("Intent not recognized.")
```
