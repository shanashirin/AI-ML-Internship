```python
from nltk.tokenize import word_tokenize
text="Artificial intelligence is changing the world."
tokens=word_tokenize(text)
print(tokens)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
text="I want to book a train ticket to Delhi. "
tokens=word_tokenize(text)
stop_words=set(stopwords.words("english"))
filtered=[]
for word in tokens:
    if word.lower() not in stop_words:
        filtered.append(word)
print(filtered)
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
print(stemmer.stem("running"))
print(stemmer.stem("reading"))
print(stemmer.stem("Learning"))
print(stemmer.stem("playing"))
```

```python

from nltk.tokenize import word_tokenize
from sympy import true
print("Ai chatbot")
print("type 'bye' to exit.")
while true:
    user=input("You:").lower()
    if user=="bye":
        print("Bot:Goodbye!")
        break
    tokens=word_tokenize(user)
    if "hello" in tokens or "hi" in tokens:
        print("Bot:Hello! How can I help you?")
    elif "fee" in tokens or "fees" in tokens:
        print("Bot:The fee for the course is $100.")
    elif "duration" in tokens or "length" in tokens:
        print("Bot:The course duration is 3 months.")
    elif "contact" in tokens or "email" in tokens:
        print("Bot:You can contact us at contact@company.com")
    else:
        print("Bot:I'm sorry, I didn't understand that. Can you please rephrase?")
```
