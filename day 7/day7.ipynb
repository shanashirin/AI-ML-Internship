```python
s=input("enter a string:")
rev=s[::-1]
if s==rev:
    print("it is palindrome")
else:
    print("it is not palindrome")
```

```python
num=[10,1,2,30,80,78]
largest_num=num[0]
for nums in num:
    if nums>largest_num:
        largest_num=nums
print("largest number is:",largest_num)
```

```python
s=input("Enter a string:")
count="aeiouAEIOU"
vowel_count=0
for i in s:
    if i in count:
        vowel_count+=1
print("number of vowels in the string is:",vowel_count)
```

```python
student=[]
for i in range(5):
    name=input("Enter name:")
    age=int(input("Enter age:"))
    marks=int(input("Enter marks:"))
    student.append({"name":name,"age":age,"marks":marks})
print(student)
```

```python
total=0
for i in student:
    total+=i["marks"]
    avg=total/len(student)
print("average marks of the students is:",avg)
```
