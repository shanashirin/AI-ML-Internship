```python
text="HELLO WORLD"
lower=text.lower()
print(lower)
```

```python
text="Python!!! is Awesome???"
clean_text=text.translate(str.maketrans('', '', '!?'))
print(clean_text)
```

```python
text="the cat is sitting on the chair"
stop_words=['the','is','on','am','are','was','were','be','been','being','have','has','had','do','does','did']
words=text.lower().split()
for word in words:
    if word in stop_words:
        print(word) 
```

```python

import re


text="AI 2025 is Amazing!!!"
text=text.lower()
text =re.sub(r'[^\w\s]', '', text)
print(text)
```
