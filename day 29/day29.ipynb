```python
import pandas as pd 
data={
    "name":["Alice","Bob","Charlie","David"],
    "department":["HR","IT","Finance","Marketing"],
    "salary":[50000,60000,55000,45000]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.sort_values("salary"))
print(df.sort_values("salary",ascending=False))
print(df.sort_values(["department","salary"]))
```

```python
group=df.groupby("department").sum()
print(group)
```

```python
data2={
    "customer":["Alice","Bob","Charlie","David"],
    "city":["New York","Los Angeles","Chicago","Houston"],
    "purchase_amount":[100,150,200,120]
}
df2=pd.DataFrame(data2)
print(df2)
```

```python
group2=df2.groupby("city").sum()
print(group2)
```

```python
data3={
    "product":["A","B","C","D"],
    "category":["Electronics","Clothing","Electronics","Clothing"],
    "price":[100,50,150,80]
}
df3=pd.DataFrame(data3)
print(df3)
group=df3.groupby("category").sum()
print(group)
```

```python
print(df.groupby("department")["salary"].mean())
print(df.groupby("department")["salary"].sum())
print(df.groupby('department')["salary"].count())
```

```python
print(df.groupby("department")["salary"].agg(["mean","min","max"]))
```

```python
print(df["department"].value_counts())
```

```python
rep_values=df["department"].value_counts()
print(rep_values)
```
