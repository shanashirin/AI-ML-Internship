```python
from PyPDF2 import PdfReader
reader=PdfReader("📘 Day 61-Chatbot.pdf")
text=""
for page in reader.pages:
    text+=page.extract_text()
print(text)
```

```python
chunk1 = "Artificial Intelligence is transforming many industries."

chunk2 = "It is widely used in healthcare, education, banking, transportation, and cybersecurity."

chunk3 = "Machine Learning is a branch of AI that enables computers to learn from data."

chunk4 = "Deep Learning uses neural networks to solve complex problems."

chunk5 = "Retrieval-Augmented Generation (RAG) combines document retrieval with Large Language Models to provide accurate answers from custom documents."

chunks = [chunk1, chunk2, chunk3, chunk4, chunk5]

for i, chunk in enumerate(chunks, start=1):
    print(f"Chunk {i}:")
    print(chunk)
    print()
```
