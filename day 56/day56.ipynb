```python
sentences=[["This", "is", "the", "first", "sentence"], ["This", "is", "the", "second", "sentence"],["This", "is", "the", "third", "sentence"], ["This", "is", "the", "fourth", "sentence"]]
```

```python
from gensim.models import Word2Vec
model=Word2Vec(sentences,vector_size=100,window=5,min_count=1)
```

```python
print(model.wv['This'])
```

```python
print(model.wv.most_similar('This'))
```
