```python
import matplotlib.pyplot as plt 
x=[1,2,3,4]
y1=[10,20,30,40]
y2=[50,90,80,70]
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y1,label='Line 1')
plt.plot(x,y2,label='Line 2')
plt.legend()
plt.figure(figsize=(8,6))
plt.show()
plt.savefig("line_plot.png")
```

```python
plt.plot(x,y1,label='Line 1')
plt.plot(x,y2,label='Line 2')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,label='Line 1')
plt.subplot(2,1,2)
plt.plot(x,y2,label='Line 2')
plt.show()
```

```python
y3=[15,25,35,45]
y4=[55,95,85,75]
plt.subplot(2,1,1)
plt.plot(x,y1,label='Line 1')
plt.subplot(2,1,2)
plt.plot(x,y2,label='Line 2')
plt.subplot(2,1,1)
plt.plot(x,y3,label='Line 3')
plt.subplot(2,1,2)
plt.plot(x,y4,label='Line 4')
plt.show()
plt.savefig("multiple_lines.png")
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 4, 9, 16]


plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Graph 1")


plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Graph 2")

plt.show()
```

```python
y3 = [1, 8, 27, 64]
y4 = [1, 16, 81, 256]
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Graph 1")
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Graph 2")
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Graph 3")
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Graph 4")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("increasing")
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title("decreasing")
plt.show()
```
