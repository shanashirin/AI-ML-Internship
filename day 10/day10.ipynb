```python
class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def display(self):
        if self.marks>=90:
            return "A"
        elif self.marks>=80:
            return "B"
        elif self.marks>=70:
            return "C"
        elif self.marks>=60:
            return "D"
        else:
            return "F"
s1=student("rohit",78)
print(s1.display())
```

```python
students=[
    student("rohit",78),
    student("sachin",85),
    student("virat",92)
]
print("students details:")
for s in students:
    print(s.name,s.marks,s.display())
topper=students[0]
for s in students:
    if s.marks>topper.marks:
        topper=s
print("Topper is",topper.name,topper.marks)
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def check_balance(self):
        print("Balance:", self.balance)

acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(300)
acc.check_balance()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def calculate_discount(self, percent):
        discount = self.price * percent / 100
        final_price = self.price - discount
        return final_price
p = Product("Laptop", 50000)
print("Final Price after discount:", p.calculate_discount(10))
```
