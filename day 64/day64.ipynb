```python
from openai import OpenAI
client=OpenAI(api_key="YOUR_API_KEY")
print("Ai chatbot")
print("Type 'bye' to exit.\n")
while True:
    user=input("You: ")
    if user.lower()=='bye':
        print("Chatbot: Goodbye!")
        break   
    response=client.response.create(
        model="gpt-4.1-mini",
        input=user
    )
    print(f"Chatbot: {response.output_text}")
```

```python
from openai import OpenAI

client = OpenAI(api_key="YOUR_API_KEY")

print("===== AI Learning Chatbot =====")
print("Ask about AI, Python, Machine Learning, or Data Science.")
print("Type 'bye' to exit.\n")

while True:

    user = input("You: ")

    if user.lower() == "bye":
        print("Bot: Goodbye!")
        break

    response = client.responses.create(
        model="gpt-4.1-mini",
        input=user
    )

    print("Bot:", response.output_text)
```
