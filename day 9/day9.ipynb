```python
students=[]
with open("students.txt","r")as file:
  for line in file:
    name,marks=line.strip().split(",")
    students.append({"name":name,"marks":int(marks)})
print(students)
```

```python
total=sum(s["marks"]for s in students)
average=total/len(students)
print(f"average marks:{average}")
```

```python
top=max(students,key=lambda s:s["marks"])
print("top:",top["name"])
```

```python
mark=list(filter(lambda s:s["marks"]>75,students))
print("students with marks above 75:",mark)
```

```python
print("total number of students:",len(students))
```

```python
lowest=min(students,key=lambda s:s["marks"])
print("lowest:",lowest["name"])
```

```python
sort=sorted(students,key=lambda s:s["marks"])
print("students sorted by marks:",sort)
```

```python
def calculate_avg(students):
    return sum(s["marks"]for s in students)/len(students)
print("average marks using function:",calculate_avg(students))
```

```python
def top(students):
    return max(students,key=lambda s:s["marks"])
print("top student using function:",top(students)["name"])
```

```python
def filter_students(students):
    return list(filter(lambda s: s["marks"] > 50, students))

print(filter_students(students))
```

```python
sort=sorted(students,key=lambda s:s["marks"],reverse=True)
second_largest=sort[1]
print("second largest:",second_largest["name"])
```

```python
passed=len(list(filter(lambda s:s["marks"]>=50,students)))
failed=len(students)-passed
print(f"passed: {passed}, failed: {failed}")
```
