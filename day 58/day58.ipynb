```python
1. Define a chatbot.

A chatbot is an Artificial Intelligence (AI) application that communicates with users through text or voice. It understands user queries, identifies their intent, and provides automated responses or performs tasks without human intervention.
```

```python
        User
          │
          ▼
    User Message
          │
          ▼
Natural Language Processing (NLP)
          │
          ▼
   Intent Recognition
          │
          ▼
Decision & Response Generator
          │
          ▼
Database / Knowledge Base
          │
          ▼
    Response to User
```

```python
3. What is NLP in chatbots?

Natural Language Processing (NLP) is a branch of AI that enables chatbots to understand, interpret, and respond to human language. It performs tasks such as tokenization, stop-word removal, lemmatization, intent detection, and entity recognition to provide accurate responses.
```

```python
Rule-Based Chatbot	AI-Based Chatbot
Uses predefined rules	Uses AI and Machine Learning
Limited responses	Flexible responses
Cannot learn from interactions	Learns from data and improves
Easy to develop	More complex to develop
Best for FAQs	Best for virtual assistants and customer support
```
