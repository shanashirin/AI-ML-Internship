```python
conversation=[]
conversation.append({
    "role":"user",
    "content":"hello"
})
conversation.append({
    "role":"assistant",
    "content":"Hello! How can I assist you today?"
})
conversation.append({
    "role":"user",
    "content":"Can you help me with Python programming?"
})
conversation.append({
    "role":"assistant",
    "content":"Of course! What specific aspect of Python programming do you need help with?"
})
conversation.append({
    "role":"user",
    "content":"I want to learn about lists and how to manipulate them."
})
```

```python
from openai import OpenAI

client = OpenAI(api_key="YOUR_API_KEY")

conversation = []

print("🤖 AI Chatbot with Memory")
print("Type 'bye' to exit.\n")

while True:

    user = input("You: ")

    if user.lower() == "bye":
        print("Bot: Goodbye!")
        break

   
    conversation.append({
        "role": "user",
        "content": user
    })

  
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=conversation
    )

    answer = response.output_text

    print("Bot:", answer)

  
    conversation.append({
        "role": "assistant",
        "content": answer
    })
```

```python
#              User
#                │
#                ▼
#       User sends message
#                │
#                ▼
#  Store message in conversation list
#                │
#                ▼
#  Send complete conversation history
#                │
#                ▼
#         OpenAI AI Model
#                │
#                ▼
#  Generate context-aware response
#                │
#                ▼
#  Display response to user
#                │
#                ▼
#  Store assistant response
#                │
#                ▼
#      Wait for next message
```
