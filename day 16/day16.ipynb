```python
students=[{"name":"shana","marks":90},{"name":"isha","marks":None},{"name":"fidha","marks":-70},{"name":"zuha","marks":50},{"name":"liya","marks":40},{"name":"rana","marks":30}]
cleaned=[]
for s in students:
    if s["marks"]is not None and s["marks"]>=0:
        cleaned.append(s)
print(cleaned)
```

```python
for s in cleaned:
    s["status"]="pass" if s["marks"]>=40 else "fail"
    s["percentage"]=s["marks"]
    if s["marks"]>=90:
        print("A")
    elif s["marks"]>=80:
        print("B")
    elif s["marks"]>=70:
        print("c")
    elif s["marks"]>=60:
        print("D")
    elif s["marks"]>=50:
        print("E")
    else:
        print("Fail")
print("transformed data",cleaned)

```

```python

cleaned.sort(key=lambda s:s["marks"],reverse=True)
print(cleaned)
print("Top 3 students:")
for s in cleaned[:3]:
    print(s)
```

```python
result={"pass":[],"fail":[]}
for s in cleaned:
    if s["marks"]>=50:
        result["pass"].append(s)
    else:
        result["fail"].append(s)
print(result)
print(len(result["pass"]))
print(len(result["fail"]))
```
