```python
import matplotlib.pyplot as plt
x=[1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]
plt.plot(x,y)
plt.title('Line Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y,linestyle='--',marker='o')
plt.grid()
plt.show()
```

```python
students=['Alice', 'Bob', 'Charlie', 'David', 'Eve']
marks=[85, 90, 78, 92, 88]
plt.bar(students, marks)
plt.title('Bar Graph')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

```python
sales=[100, 150, 200, 250, 300]
months=['Jan', 'Feb', 'Mar', 'Apr', 'May']
plt.bar(sales,months)
plt.title('sales graph')
plt.xlabel('sales')
plt.ylabel('months')
plt.show()
```

```python
product=['A', 'B', 'C', 'D', 'E']
quantity=[10, 20, 15, 25, 30]
plt.bar(product,quantity)
plt.title('Product Quantity')
plt.xlabel('Product')
plt.ylabel('Quantity')
plt.show()
```

```python
departments=['HR', 'Finance', 'IT', 'Marketing', 'Sales']
employees=[5, 8, 12, 6, 10]
plt.pie(employees,labels=departments)
plt.title('Employees in Departments')
plt.show()
```

```python
expenses=[500, 300, 200, 100]
categories=['Rent', 'Food', 'Transport', 'Entertainment']
plt.pie(expenses,labels=categories)
plt.title('Monthly Expenses')
plt.show()
```

```python
course=['Python', 'Data Science', 'Machine Learning', 'Deep Learning']
popularity=[80, 90, 85, 70]
plt.pie(popularity,labels=course)
plt.title('Course Popularity')
plt.show()
```

```python
height=[150, 160, 170, 180, 190]
weight=[50, 60, 70, 80, 90]
plt.scatter(height, weight)
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
hours=[1, 2, 3, 4, 5]
marks=[10, 20, 30, 40, 50]
plt.scatter(hours, marks)
plt.title('Hours Studied vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.show()
```

```python
temperature=[30, 32, 28, 25, 27]
sales=[200, 220, 180, 150, 170]
plt.scatter(temperature, sales)
plt.title('Temperature vs Sales')
plt.xlabel('Temperature (°C)')
plt.ylabel('Sales')
plt.show()
```
