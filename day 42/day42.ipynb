```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
data={
    "Student hours":[1,2,3,4,5],
    "Result":[0,1,1,1,0]
}
df=pd.DataFrame(data)
print(df)
X=df[["Student hours"]]
y=df[["Result"]]
model=GaussianNB()
model.fit(X,y)
print(model.predict([[6]]))
```

```python
data={
    "disease":[0,1,0,1,0],
    "symptom":[0,1,1,1,0]
}
df=pd.DataFrame(data)
print(df)
X=df[["symptom"]]
y=df[["disease"]]
model=GaussianNB()
model.fit(X,y)
print(model.predict([[1]]))
```

```python
data={
    "salary":[1000,2000,3000,4000,5000],
    "spam":[0,1,1,1,0]
}
df=pd.DataFrame(data)
print(df)
X=df[["salary"]]
y=df[["spam"]]
model=GaussianNB()
model.fit(X,y)
print(model.predict([[6000]]))

```

```python
# Given:

# Total Emails = 200
# Spam Emails = 150
# 1. Probability of Spam

# P(Spam)=
# 200
# 150
# 	​

# =0.75

# Answer: 0.75 or 75%

# 2. Probability of Not Spam

# P(Not Spam)=
# 200
# 50
# 	​

# =0.25

# Answer: 0.25 or 25%
```
