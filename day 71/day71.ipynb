```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("📘 Day 71-RAG.pdf")
documents=loader.load()
print(documents[:2])
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(documents)

print("Total Chunks:", len(chunks))
print(chunks[0].page_content)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

# Create embedding model
embedding = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

# Store chunks in ChromaDB
db = Chroma.from_documents(
    chunks,
    embedding
)

# Question 1
docs1 = db.similarity_search(
    "What is the admission fee?",
    k=3
)

print("Question 1 Results")
for doc in docs1:
    print(doc.page_content)
    print("-" * 50)

# Question 2
docs2 = db.similarity_search(
    "Who is eligible for admission?",
    k=3
)

print("Question 2 Results")
for doc in docs2:
    print(doc.page_content)
    print("-" * 50)

# Question 3
docs3 = db.similarity_search(
    "What is the internship duration?",
    k=3
)

print("Question 3 Results")
for doc in docs3:
    print(doc.page_content)
    print("-" * 50)
```
