```python
print("Ai Assistant")
print("type 'Bye' to exit \n")
while True:
    user=input("You:").lower()
    if user=="Hello":
        print("Bot: Hello! How can I assist you today?")
    elif user=="Hi":
        print("Bot: Hi there! What can I do for you?")
    elif user=="How are you?":
        print("Bot: I'm just a program, but I'm functioning as expected! How can I help you?")
    elif user=="What is your name?":
        print("Bot: I am an AI Assistant. You can call me whatever you like!")
    elif user=="Bye":
        print("Bot: Goodbye! Have a great day!")
        break
    else:
        print("Bot: I'm sorry, I don't understand that. Can you please rephrase?")
```

```python
responses = {
    "hello": "Hello! How can I assist you today?",
    "hi": "Hi there! What can I do for you?",
    "how are you?": "I'm just a program, but I'm functioning as expected! How can I help you?",
    "what is your name?": "I am an AI Assistant. You can call me whatever you like!",
    "bye": "Goodbye! Have a great day!"
}

user = input("You: ").lower()
print(responses.get(user, "I'm sorry, I don't understand that. Can you please rephrase?"))
```

```python
print("===== College Enquiry Chatbot =====")
print("Type 'bye' to exit.\n")

while True:

    user = input("You: ").lower()

    if user == "fees":
        print("Bot: The course fee is ₹15,000.")

    elif user == "duration":
        print("Bot: The course duration is 3 months.")

    elif user == "location":
        print("Bot: Our college is located in Kochi.")

    elif user == "admission":
        print("Bot: Admissions are open now.")

    elif user == "contact":
        print("Bot: Contact us at 9876543210.")

    elif user == "bye":
        print("Bot: Thank you for visiting.")
        break

    else:
        print("Bot: Sorry! I don't understand your question.")
```

```python
responses = {
    "hello": "Hello! How can I help you?",
    "hi": "Hi! Welcome!",
    "your name": "I am AI Chatbot.",
    "fees": "Course fee is ₹15,000.",
    "duration": "Course duration is 3 months.",
    "location": "We are located in Kochi.",
    "admission": "Admissions are open now.",
    "contact": "Call us at 9876543210."
}

print("===== AI Chatbot =====")

while True:

    user = input("You: ").lower()

    if user == "bye":
        print("Bot: Goodbye!")
        break

    print("Bot:", responses.get(user, "Sorry! I don't understand your question."))
```

```python
faq = {
    "hello": "Hello! Welcome.",
    "hi": "Hi! Nice to meet you.",
    "fees": "Course fee is ₹15,000.",
    "duration": "Course duration is 3 months.",
    "location": "We are located in Kochi.",
    "admission": "Admissions are open now.",
    "contact": "Call us at 9876543210."
}

print("===== AI Chatbot =====")

while True:

    user = input("You: ").lower()

    if user == "bye":
        print("Bot: Goodbye!")
        break

    found = False

    for key in faq:

        if key in user:
            print("Bot:", faq[key])
            found = True
            break

    if not found:
        print("Bot: Sorry! I don't understand your question.")
```
