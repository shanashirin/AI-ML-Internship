```python
import pandas as pd 
from sklearn.svm import SVC
data={"disease":[1,1,1,1,0,0,0,0],"age":[20,30,40,50,20,30,40,50]}
df=pd.DataFrame(data)
X=df[["disease"]]
y=df["age"]
model=SVC()
model.fit(X,y)
print(model.predict([[1]]))
```

```python
data={
    "student":[1,2,3,4,5,6,7,8],
    "result":[1,1,1,1,0,0,0,0]
}
df=pd.DataFrame(data)
X=df[["student"]]
y=df["result"]
model=SVC()
model.fit(X,y)
print(model.predict([[1]]))
```

```python
data={
    "spam":[1,1,1,1,0,0,0,0],
    "email":[20,30,40,50,20,30,40,50]
}
df=pd.DataFrame(data)
X=df[["spam"]]
y=df["email"]
model=SVC()
model.fit(X,y)
print(model.predict([[1]]))
```

```python
# | Algorithm                        | Type                        | Advantages                                        | Disadvantages                               | Best Use Cases                              |
# | -------------------------------- | --------------------------- | ------------------------------------------------- | ------------------------------------------- | ------------------------------------------- |
# | **Logistic Regression**          | Classification              | Simple, fast, easy to interpret                   | Performs poorly on complex non-linear data  | Spam detection, medical diagnosis           |
# | **K-Nearest Neighbors (KNN)**    | Classification & Regression | Easy to understand, no training phase             | Slow for large datasets, sensitive to noise | Pattern recognition, recommendation systems |
# | **Decision Tree**                | Classification & Regression | Easy visualization, handles non-linear data       | Can overfit easily                          | Customer segmentation, decision making      |
# | **Random Forest**                | Ensemble Learning           | High accuracy, reduces overfitting                | More computationally expensive              | Fraud detection, stock prediction           |
# | **Support Vector Machine (SVM)** | Classification & Regression | Effective in high-dimensional data, good accuracy | Slow on large datasets, harder to interpret | Image recognition, text classification      |
```
