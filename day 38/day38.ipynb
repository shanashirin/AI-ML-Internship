```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
data={
    "marks":[10,70,90,80,40,30],
    "result":[0,1,1,1,0,0]
}
df=pd.DataFrame(data)
X=df[["marks"]]
y=df["result"]
model=LogisticRegression()
model.fit(X,y)
pred=model.predict([[50],[60],[20]])
print(pred)
```

```python
data={
    "disease":[0,1,1,0,1,0],
    "symptom":[0,1,1,0,1,0]
}
df=pd.DataFrame(data)
X=df[["symptom"]]
y=df["disease"]
model=LogisticRegression()
model.fit(X,y)
pred=model.predict([[0],[1]])
print(pred)
```

```python
data={
    "spam":[0,1,1,0,1,0],
    "message":[90,80,40,30,70,20]
}
df=pd.DataFrame(data)
X=df[["message"]]
y=df["spam"]
model=LogisticRegression()
model.fit(X,y)
pred=model.predict([[50],[60],[20]])
print(pred)
```

```python
model.predict_proba([[50],[60],[20]])
print(model.predict_proba([[50],[60],[20]]))
#Meaning:

# 15% probability of Fail
# 85% probability of Pass

# The highest probability becomes the final prediction.
```

```python
# #| Probability | Result |
# | ----------- | ------ |
# | 0.20        | Fail   |
# | 0.40        | Fail   |
# | 0.80        | Pass   |
# | 0.95        | Pass   |
```

```python
# | Problem                | Logistic Regression Can Be Used? |
# | ---------------------- | -------------------------------- |
# | Pass/Fail Prediction   | Yes                              |
# | House Price Prediction | No                               |
# | Spam Detection         | Yes                              |
# | Disease Prediction     | Yes                              |
# | Salary Prediction      | No                               |
```
