```python
import numpy as np
arr=np.array([10,70,40,60,20])
print(arr[0])
print(arr[-1])
print(arr[1:4])
```

```python
arr2=np.array([[10,20,30],[40,50,60],[70,800,90]])
print(arr2[0,1])
print(arr2[0])
print(arr2[1])
print(arr2[:,1])
```

```python
arr2=np.array([[[10,20,30],[40,50,60]],[[70,800,90],[100,110,120]],[[130,140,150],[160,170,180]]])
print(arr2[0])
print(arr2[0][1])
print(arr2[0][1][2])
```
