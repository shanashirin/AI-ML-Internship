```python
import chromadb

client = chromadb.Client()

collection = client.create_collection("internship_docs")

documents = [
    "The internship fee is Rs. 5000.",
    "The internship duration is 3 months.",
    "Placement assistance is available.",
    "Certificate will be provided after completion.",
    "Python is taught during the internship.",
    "Machine Learning is included in the syllabus.",
    "Deep Learning is covered in Module 10.",
    "Students receive hands-on projects.",
    "Classes are conducted online.",
    "The internship starts on August 1."
]

ids = [str(i) for i in range(len(documents))]

collection.add(
    documents=documents,
    ids=ids
)

print("Documents stored successfully.")
```

```python
questions = [
    "What is the internship fee?",
    "How long is the internship?",
    "Is placement assistance available?"
]

for q in questions:
    print("\nQuestion:", q)

    results = collection.query(
        query_texts=[q],
        n_results=3
    )

    print(results["documents"])
```

```python
query = "How long is the internship?"

for n in [1, 3, 5]:

    print(f"\nTop {n} Results")

    results = collection.query(
        query_texts=[query],
        n_results=n
    )

    print(results["documents"])
```

```python
import chromadb

client = chromadb.Client()

collection = client.get_collection("internship_docs")

query = input("Enter your question: ")

results = collection.query(
    query_texts=[query],
    n_results=3
)

print("\nTop 3 Similar Documents:\n")

for i, doc in enumerate(results["documents"][0], start=1):
    print(f"{i}. {doc}")
```
