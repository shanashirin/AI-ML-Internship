```python
s=input("Enter a string:")
rev=s[::-1]
if s==rev:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
students=[{"name":"Alice","marks":85},{"name":"Bob","marks":80}]
top=students[0]
for s in students:
    if s["marks"]>top["marks"]:
        top=s
print(top)
```
