```python
text=input("Enter the string:")
vowels="aeiouAEIOU"
vowel_count=0
consonant_count=0
for ch in text:
    if ch in vowels:
        vowel_count+=1
    elif ch.isalpha():
        consonant_count+=1
print("Number of vowels in the string is:",vowel_count)
print("Number of consonants in the string is:",consonant_count)
```

```python
nums = [1,2,3,4,2,3,4,5]
freq = {}

for num in nums:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print("Frequency of each number in the list is:", freq)
```
