```python
def num(n):
    for num in range(n):
        yield num
for num in num(5):
    print(num)
```

```python
def num(n):
    for i in range(n):
        if i%2==0:
            yield i
for num in num(10):
    print("even num:",num)
```

```python
def fib(n):
    a,b=0,1
    for _ in range(n):
        yield a
        a,b=b,a+b
for num in fib(10):
    print(num)
```
