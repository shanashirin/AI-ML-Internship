```python
from sklearn.linear_model import LinearRegression
import pandas as pd
data={
    "hours":[1,2,3,4],
    "marks":[90,80,40,30]
}
df=pd.DataFrame(data)
X=df[["hours"]]
y=df[["marks"]]
model=LinearRegression()
model.fit(X,y)
prediction=model.predict([[6]])
print(prediction)
```

```python
data={
    "salary":[1000,5000,4000,7000],
    "employee":[8,4,2,3]
}
df=pd.DataFrame(data)
X=df[["salary"]]
y=df[["employee"]]
model=LinearRegression()
model.fit(X,y)
prediciton=model.predict([[7]])
print(prediciton)
```

```python
data={
    "product":[1000,5000,4000,7000],
    "sales":[8,4,2,3]
}
df=pd.DataFrame(data)
X=df[["product"]]
y=df[["sales"]]
model=LinearRegression()
model.fit(X,y)
prediciton=model.predict([[6]])
print(prediciton)
```

```python
from sklearn.linear_model import LogisticRegression
data={
    "hours":[1,2,3,4,5],
    "result":[0,0,0,1,1]
}
df=pd.DataFrame(data)
X=df[["hours"]]
y=df[["result"]]
model=LogisticRegression()
model.fit(X,y)
prediction=model.predict([[5]])
print(prediction)
```

```python
data={
    "message_length":[10,20,30,40,50],
    "spam":[0,0,0,1,1]
}
df=pd.DataFrame(data)
print(df)
X=df[["message_length"]]
y=df[["spam"]]
model=LogisticRegression()
model.fit(X,y)
prediction=model.predict([[60]])
print(prediction)
```

```python
data={
    "fever":[1,1,1,0,0],
    "disease":[1,1,1,0,0]
}
df=pd.DataFrame(data)
X=df[["fever"]]
y=df[["disease"]]
model=LogisticRegression()
model.fit(X,y)
prediction=model.predict([[1]])
print(prediction)
```
